# Visualization Project 

## The Data

This classic dataset contains the prices and other attributes of almost 54,000 diamonds.

#### Content

* price price in US dollars (\$326--\$18,823)

* carat weight of the diamond (0.2--5.01)

* cut quality of the cut (Fair, Good, Very Good, Premium, Ideal)

* color diamond colour, from J (worst) to D (best)

* clarity a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

* x length in mm (0--10.74)

* y width in mm (0--58.9)

* z depth in mm (0--31.8)
 
* depth total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)

* table width of top of diamond relative to widest point (43--95)

## Imports and Data

```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
diamonds = pd.read_csv('./diamonds.csv')
```

```python
diamonds.head()
```

```python
diamonds.info()
```

#### A scatterplot of price versus carat

```python
sns.scatterplot(data=diamonds, x='carat', y="price")
```

#### Changing the alpha  and edgecolor parameters to deal with the overlapping and white edgemarker issue.

```python
sns.scatterplot(data=diamonds, x='carat', y="price", alpha=.2, edgecolor='lightblue')
```

#### Plot Resize

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=diamonds, x='carat', y="price", alpha=.2, edgecolor='lightblue')
```

#### A histogram of the price column

```python
plt.figure(figsize=(10,8))
sns.distplot(diamonds['price'],kde=None)
```

#### A count plot of the instances per cut type

```python
sns.countplot(diamonds['cut'])
```

#### The price distribution per cut type

```python
sns.boxplot(data=diamonds,y="price", x="cut")
```

#### Changing the cut type ordering

```python
sns.boxplot(data=diamonds,y="price", x="cut", order=["Fair","Ideal","Good", "Very Good", "Premium"], palette="deep")
```

#### Changing the color

```python
sns.boxplot(data=diamonds,y="price", x="cut", order=["Fair","Ideal","Good", "Very Good", "Premium"], palette="cool" )
```
