```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv('test.csv')
```

```python
df1_2 = df1.drop('Cabin',axis=1)
```

```python
df1_3 = df1_2
sex = pd.get_dummies(df1_3['Sex'],drop_first=True)
embark = pd.get_dummies(df1_3['Embarked'],drop_first=True)
```

```python
df1_4 = df1_3.drop(["Sex","Embarked"], axis=1)
```

```python
df1_5 = pd.concat([df1_4,sex,embark],axis=1)
```

```python
df1_5.head()
```

```python
def isMrs(x):
    if 'Mrs' in x:
        return 1
    else:
        return 0

df1_5["Mrs"] = df1_5['Name'].apply(isMrs)
```

```python
def ticketLen(x):
    
    return len(x)

df1_5["ticketLen"] = df1_5['Ticket'].apply(ticketLen)
```

```python
df1_5.head()
```

```python
df1_6 = df1_5.drop(['Name','Ticket',"PassengerId"], axis=1)
```

```python
df1_6
```

```python
def fillMissingAge(cols):
    pClass = cols[0]
    age = cols[1]
    
    if pd.isnull(age):
        if pClass == 1:
            return 38.2
        elif pClass == 2:
            return 29.9
        else:
            return 25.1
    else:
        return age

df1_6['Age'] = df1_6[['Pclass',"Age"]].apply(fillMissingAge,axis=1)
```

```python
df1_6.head()
```

```python
pclass = pd.get_dummies(df1_6['Pclass'])
```

```python
df1_6.drop(["Pclass"],axis=1)
```

```python
df1_7 = pd.concat([df1_6,pclass],axis=1)
```

```python
df1_7
```

```python
df1_8 = df1_7[["Age", "SibSp", "Parch", "Fare", "Mrs","ticketLen","male","Q","S",1,2,3]]
```

```python
df1_8.head()
```
